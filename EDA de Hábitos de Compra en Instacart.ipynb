### Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
agrupado = order.groupby('product_id')['reordered'].agg(['sum', 'count']).reset_index()
agrupado.columns = ['product_id', 'total_reordered', 'total_orders']

agrupado['reorder_ratio'] = agrupado['total_reordered'] / agrupado['total_orders']


agrupado = agrupado.merge(product[['product_id', 'product_name']], on='product_id')

filtrado = agrupado[agrupado['total_orders'] >= 100]

top_reordered = filtrado.sort_values(by='reorder_ratio', ascending=False).head(20)

plt.figure(figsize=(10,6))
plt.hist(filtrado['reorder_ratio'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Proporción de reordenamiento')
plt.ylabel('Número de productos')
plt.title('Distribución de proporciones de reordenamiento (productos con ≥ 100 pedidos)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

**Observaciones**

Como podemos ver, existe una baja proporción de reordenamiento en los productos, en otras palabras, el consumo es bajo ya que no son comprados de forma recurrente. Además, se puede observar que existen productos que tienen una decisión de recompra en los pedidos generados después de la primera compra (Se determina por que tan cerca está el reordenamiento al valor de 1.0).

```python
order_prod_users = order.merge(cart[['order_id', 'user_id']], on='order_id')

user_total_orders = order_prod_users.groupby('user_id')['reordered'].count()
user_total_reorders = order_prod_users.groupby('user_id')['reordered'].sum()

user_reorder_rate = (user_total_reorders / user_total_orders).reset_index()
user_reorder_rate.columns = ['user_id', 'reorder_rate']

plt.figure(figsize=(10,6))
plt.hist(user_reorder_rate['reorder_rate'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribución de la proporción de productos reordenados por usuario')
plt.xlabel('Proporción de productos reordenados')
plt.ylabel('Número de usuarios')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

**Observaciones**

Muchos clientes tienen ratios cercanos a 0.5 y 0.8, lo cual indica que hay comportamiento de re compra.
